```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x, y)
plt.title('Dummy')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
fig = plt.figure()
# Add an axes at position *rect* [*left*, *bottom*, *width*,*height*] where all quantities are in fractions of figure
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(x, y)

axes1.set_title('Bigger plot')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')

axes2.set_title('Smaller plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].set_title('First')
axes[0].plot(x, y)
axes[1].set_title('Second')
axes[1].plot(x, y)
```

## Figure size and DPI

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1 ,figsize = (8, 2))
plt.tight_layout()
axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
fig.savefig('my_picture.png', dpi =200)
```

```python
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.plot(y, x)
ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_ylabel('X Label')
```

###                                      Colouring

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = 'green', linewidth = 3, alpha = 0.5)  # RGB hexcode can also be used.
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = 'green',lw = 3,  linestyle = ':', marker = 'o')  # linestyle = ls
```
